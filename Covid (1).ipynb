```python
# GROUP BY
#Question 1: Enrollment Analysis by Phase
enrollment_by_phase = df.groupby('Phases')['Enrollment'].sum()
print(enrollment_by_phase)
```

```python
#Question 2: Enrollment Analysis by Age
enrollment_by_age = df.groupby('Age')['Enrollment'].sum()
enrollment_df = enrollment_by_age.reset_index()
pd.set_option('display.max_rows', None)
print(enrollment_df)
```

```python
#Question 2: Enrollment Analysis by Condition
enrollment_by_condition = df.groupby('Conditions')['Enrollment'].sum()
enrollment_df = enrollment_by_condition.reset_index()
pd.set_option('display.max_rows', None)
print(enrollment_df)
```

```python
#Question 4:  Explore the count of outcome measures for different conditions.
outcome_measures_by_conditions = df.groupby('Conditions')['Outcome Measures'].count()
print(outcome_measures_by_conditions)
```

```python
#Question 5: Explore the distribution of genders in different trial statuses.
gender_distribution_by_status = df.groupby(['Status', 'Gender'])['NCT Number'].count()
print(gender_distribution_by_status)
```
